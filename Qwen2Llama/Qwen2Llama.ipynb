## 复现qwen2.5碰到的问题：
- llama3和qwen2.5的attention_norm和ffn_norm顺序不一样，llama3是先进行attention_norm再ffn_norm，而qwen2.5则相反
- llama3和qwen2.5都有rotary_pos_emb在attention层中，但是qwen2.5的rotary_emb会得到cos和sin,然后再rotary_pos_emb函数对query和key做pos_emb
llama3中的rotary_pos_emb会先计算一个包含基础频率的复数矩阵(end, dim//2)，然后再把q,k转为复数形式。而qwen2.5则会使用分开的cos和sin张量来实现旋转操作，通过将输入张量的后一半维度取负值并拼接在前一半维度之前。复数形式更简洁高效，而通过正余弦更加直观和易于理解。
- llama3和qwen2.5的ffn隐藏层的维度确定方法不同
qwen2.5通过直接设置intermediate_size而llama3是通过设置嵌入维度dim,multiple_of和ffn_dim_mulitplier来确定。具体的计算方式如下：
```python
def compute_dim(hidden_dim, multiple_of, ffn_dim_multiplier):  
    hidden_dim = int(2 * hidden_dim / 3)   
    hidden_dim = int(ffn_dim_multiplier * hidden_dim)   
    hidden_dim = multiple_of * ((hidden_dim + multiple_of - 1) // multiple_of)   
    return hidden_dim
```
- llama3中ffn层中的命名w1,w2,w3顺序稍有相反，而qwen2中比较清晰：gate_proj, up_proj, down_proj对应到llama3中就是w1,w3,w2
其他的结构实现上大同小异，主要的区别点就在以上几点。
- llama3中q,k,v都没有设置bias，而qwen2.5中都设置了。

## DynamicCache 类
DynamicCache 类在 Hugging Face 的 Transformers 库中主要用于在生成文本时动态地存储和管理注意力机制（如自注意力和交叉注意力）的键（Key）和值（Value）缓存。这些缓存用于加速生成过程，特别是在使用束搜索（Beam Search）等复杂生成策略时。以下是 DynamicCache 类的主要作用和功能：

主要作用
动态缓存管理：

DynamicCache 类允许缓存动态增长，随着生成的令牌数量增加而扩展。
这对于生成长文本序列特别有用，因为它可以避免预先分配大量内存。
存储键和值：

缓存存储每个层的键和值张量，形状为 [batch_size, num_heads, seq_len, head_dim]。
这些张量用于在生成过程中快速访问和更新注意力机制的状态。
支持多种生成策略：

适用于贪婪解码、束搜索、采样等多种生成策略。
通过动态缓存，可以高效地管理不同策略所需的键和值。
主要方法
初始化：

__init__ 方法初始化缓存，包括键和值的列表。
检查 batch_size 和 max_cache_len 是否大于 0，确保参数有效。
更新缓存：

update 方法用于更新缓存中的键和值。
根据层索引 layer_idx，将新的键和值张量添加到缓存中，或者与现有张量拼接。
获取缓存信息：

get_seq_length 方法返回缓存中某个层的序列长度。
get_max_cache_shape 方法返回缓存的最大序列长度（对于 DynamicCache，返回 None，表示没有固定的最大长度）。
批量操作：

batch_repeat_interleave 方法用于在批量维度上重复缓存。
batch_select_indices 方法用于选择缓存中特定的批量索引。
序列操作：

crop 方法用于裁剪缓存到指定的最大长度。
reset 方法用于重置缓存，将所有键和值张量清零。

## StaticCache 类
StaticCache 类在 Hugging Face 的 Transformers 库中用于在生成文本时管理注意力机制（如自注意力和交叉注意力）的键（Key）和值（Value）缓存。与 DynamicCache 不同，StaticCache 类具有固定的最大序列长度，这意味着它在初始化时需要指定最大缓存长度，并且在生成过程中不会动态增长。以下是 StaticCache 类的主要作用和功能：

主要作用
固定缓存管理：

StaticCache 类在初始化时需要指定最大序列长度 max_cache_len。
这意味着缓存的大小是固定的，不会随着生成的令牌数量增加而扩展。
存储键和值：

缓存存储每个层的键和值张量，形状为 [batch_size, num_heads, seq_len, head_dim]。
这些张量用于在生成过程中快速访问和更新注意力机制的状态。
支持多种生成策略：

适用于贪婪解码、束搜索、采样等多种生成策略。
通过固定缓存，可以高效地管理不同策略所需的键和值。
主要方法
初始化：

__init__ 方法初始化缓存，包括键和值的列表，并检查 batch_size 和 max_cache_len 是否大于 0。
初始化时需要指定 max_cache_len，表示缓存的最大序列长度。
更新缓存：

update 方法用于更新缓存中的键和值。
根据层索引 layer_idx，将新的键和值张量添加到缓存中，或者与现有张量拼接。
如果缓存已满，则会抛出错误或进行适当的处理。
获取缓存信息：

get_seq_length 方法返回缓存中某个层的序列长度。
get_max_cache_shape 方法返回缓存的最大序列长度。
批量操作：

batch_repeat_interleave 方法用于在批量维度上重复缓存。
batch_select_indices 方法用于选择缓存中特定的批量索引。
序列操作：

crop 方法用于裁剪缓存到指定的最大长度。
reset 方法用于重置缓存，将所有键和值张量清零。


# AttentionMaskConverter类
AttentionMaskConverter 类通过提供多种方法来处理和转换注意力掩码，确保在不同的生成和推理场景中能够高效地管理注意力机制。它支持创建因果掩码、转换 2D 掩码到 4D 掩码，并处理滑动窗口机制，从而优化模型的性能和灵活性。以下是 AttentionMaskConverter 类的主要功能总结：

创建因果掩码：生成一个因果掩码，确保模型只能看到当前和之前的位置。
转换 2D 掩码到 4D 掩码：将 2D 掩码转换为 4D 掩码，以便与自注意力机制中的键和值张量兼容。
处理滑动窗口：支持滑动窗口机制，限制模型在生成时只能看到有限的上下文。
忽略因果掩码：在某些情况下，可以忽略因果掩码，直接使用 is_causal 参数，优化性能。
处理未关注的令牌：确保所有令牌都能被正确处理，特别是在使用 scaled_dot_product_attention 时。
通过这些功能，AttentionMaskConverter 类在生成文本时提供了灵活且高效的注意力掩码管理。
